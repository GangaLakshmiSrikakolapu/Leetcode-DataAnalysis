# Data_cleaning
### Purpose:
- Clean dataset and prepare for analysis.

```python
import pandas as pd

df = pd.read_csv("../data/raw/leetcode_dataset - lc.csv")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["companies"] = df["companies"].fillna("Unknown")
```

```python
df["related_topics"] = df["related_topics"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.to_csv("../data/processed/cleaned_leetcode.csv",
          index=False)
```

### Theory
- Remove duplicate records.
- Fill missing company names.
- Fill missing topics.
- Create clean dataset.
- Save processed data.
