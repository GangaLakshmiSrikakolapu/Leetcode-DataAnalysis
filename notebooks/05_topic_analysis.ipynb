# Topic_Analysis
### Purpose:
- Analyze most frequent interview topics.

```python
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_leetcode.csv")
```

## Top Companies

```python
companies = (
    df["companies"]
    .str.split(",")
    .explode()
)

companies.value_counts().head(20)
```

## Top Topics

```python
topics = (
    df["related_topics"]
    .str.split(",")
    .explode()
)

topics.value_counts().head(20)
```

## Visualization

```python
topics.value_counts().head(10).plot(
    kind="bar",
    figsize=(10,5)
)
```

### Theory
- Identify common topics.
- Discover frequently asked concepts.
- Analyze company trends.
- Understand interview focus.
- Build topic priorities.
