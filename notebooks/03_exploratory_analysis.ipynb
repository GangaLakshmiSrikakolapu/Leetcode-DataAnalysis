# Exploratory_Analysis
### Purpose:
- Perform overall EDA.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/cleaned_leetcode.csv")
```

## Difficulty Distribution

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df,
              x="difficulty")

plt.title("Difficulty Distribution")
plt.show()
```

## Likes vs Dislikes

```python
plt.figure(figsize=(10,5))

sns.scatterplot(
    data=df,
    x="likes",
    y="dislikes"
)

plt.show()
```

## Correlation

```python
numeric_cols = [
    "acceptance_rate",
    "frequency",
    "likes",
    "dislikes",
    "rating"
]

corr = df[numeric_cols].corr()

plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.show()
```
