# Difficulty_analysis.
### Purpose:
- Analyze Easy, Medium and Hard problems.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/cleaned_leetcode.csv")
```

## Difficulty Count

```python
difficulty_count = df["difficulty"].value_counts()

difficulty_count
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x=difficulty_count.index,
    y=difficulty_count.values
)

plt.title("Problems by Difficulty")

plt.show()
```

## Average Acceptance

```python
df.groupby("difficulty")["acceptance_rate"].mean()
```
